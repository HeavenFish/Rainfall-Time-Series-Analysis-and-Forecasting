```python
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import seaborn as sns
import folium as fol
import numpy as np
import os
```

```python
#SNSI,DSKT,TAPE,SBRS,KUAM 
def load_data(name):
    path = rf"D:\my_project\Rainfall\weather\2018-20210912T153840Z-001\2018\data_weather_01\{name}.2018.01.csv"
    df = pd.read_csv(path) 
    df = pd.DataFrame(df) 
    for i in range(2,13):
        if(i<10):
            df2 = pd.read_csv(rf"D:\my_project\Rainfall\weather\2018-20210912T153840Z-001\2018\data_weather_0{i}\{name}.2018.0{i}.csv")
        else:
            df2 = pd.read_csv(rf"D:\my_project\Rainfall\weather\2018-20210912T153840Z-001\2018\data_weather_{i}\{name}.2018.{i}.csv")
        #print(df2)
        df = pd.concat([df, df2], ignore_index=True)
    return df

def create_date(df):
    date = []
    for i in range(len(df)):
        if (len(str(df['day'].iloc[i])) == 1 and len(str(df['month'].iloc[i])) == 1):
            date.append(str(df['time'].iloc[i]) + ' ' + '0' + str(df['day'].iloc[i]) + '-' + '0' + str(df['month'].iloc[i]) + '-' + str(df['year'].iloc[i]))
        elif len(str(df['day'].iloc[i])) == 1:
            date.append(str(df['time'].iloc[i]) + ' ' + '0' + str(df['day'].iloc[i]) + '-' + str(df['month'].iloc[i]) + '-' + str(df['year'].iloc[i]))
        elif len(str(df['month'].iloc[i])) == 1:
            date.append(str(df['time'].iloc[i]) + ' ' + str(df['day'].iloc[i]) + '-' + '0' + str(df['month'].iloc[i]) + '-' + str(df['year'].iloc[i])) 
        else:
            date.append(str(df['time'].iloc[i]) + ' ' + str(df['day'].iloc[i]) + '-' + str(df['month'].iloc[i]) + '-' + str(df['year'].iloc[i]))
    df['date'] = date
    return df

def percent_missing(name, df):
    missing = (df[name].mask(df[name]==-999).isna().sum())
    All = len(df)
    print(f"percent missing of {name} is ", round(missing*100/All, 2),"%")
```

```python
df_DSKT = load_data("DSKT")
df_DSKT = create_date(df_DSKT)
```
